```python

df = pd.read_csv(csv_file_path)

def sorted_table(column, ascending): 
    df_unique = df.drop_duplicates(subset='Beverage')

# Sort the DataFrame based on 'Calories' in ascending order
    sorted_df = df_unique.sort_values(by=column, ascending=ascending)


    new_df = pd.DataFrame({
        column: sorted_df['Beverage'] + ' (' + sorted_df[column].astype(str) + ')'
    })

    return new_df

table_format = 'grid'  

cal_df = sorted_table('Calories', True).head(10)
pro_df = sorted_table(' Protein (g) ', False).head(10)
chol_df = sorted_table('Cholesterol (mg)', True).head(10)
sug_df = sorted_table(' Sugars (g)', True).head(10)

sug_array = np.array(sug_df.values.flatten())
cal_array = np.array(cal_df.values.flatten())
pro_array = np.array(pro_df.values.flatten())
chol_array = np.array(chol_df.values.flatten())
rank_array = np.arange(1, 11, 1, dtype=np.int32)
combined_df = pd.DataFrame({'Rank' : rank_array, 'Calories': cal_array, ' Protein (g) ': pro_array, 'Cholesterol (mg)': chol_array,' Sugars (g)': sug_array})
print("STARBUCKS HEALTHIEST DRINKS")
print(tabulate(combined_df, headers=['Rank', 'Calories', ' Protein (g) ', 'Cholesterol (mg)', ' Sugars (g)'], tablefmt=table_format, showindex=False))

```

```python
labels = df['Beverage_category'].unique()
counts = df['Beverage_category'].value_counts()

# Convert unique counts to a NumPy array
sizes = counts.to_numpy()

labels = labels.tolist()
plt.figure(figsize=(7, 7))

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=45)

# Equal aspect ratio ensures that the pie chart is circular.
plt.axis('equal')

# Display the pie chart
plt.show()
```

```python


def sorted_table(column, ascending): 
    df_unique = df.drop_duplicates(subset='Beverage')

    # Sort the DataFrame based on the specified column
    sorted_df = df_unique.sort_values(by=column, ascending=ascending)

    new_df = pd.DataFrame({
        column: sorted_df['Beverage'] + ' (' + sorted_df[column].astype(str) + ')'
    })

    return new_df

df['Vitamins'] = pd.to_numeric(df['Vitamin A (% DV) '].str.strip().str.rstrip('%')) + pd.to_numeric(df['Vitamin C (% DV)'].str.strip().str.rstrip('%'))


cal_df = sorted_table('Calories', True).head(10)
pro_df = sorted_table(' Protein (g) ', False).head(10)
chol_df = sorted_table('Cholesterol (mg)', True).head(10)
sug_df = sorted_table(' Sugars (g)', True).head(10)
vit_df = sorted_table('Vitamins', False).head(10)

# Extract arrays

sug_array = np.array(sug_df.values.flatten())
cal_array = np.array(cal_df.values.flatten())
pro_array = np.array(pro_df.values.flatten())
chol_array = np.array(chol_df.values.flatten())
rank_array = np.arange(1, 11, 1, dtype=np.int32)
vit_array = np.array(vit_df.values.flatten())


# Create combined DataFrame
combined_df = pd.DataFrame({
    'Rank': rank_array,
    'Calories': cal_array,
    ' Protein (g) ': pro_array,
    'Cholesterol (mg)': chol_array,
    ' Sugars (g)': sug_array,
    'Vitamins': vit_array
})

# Display the combined DataFrame
table_format = 'grid'
print("STARBUCKS HEALTHIEST DRINKS")
print(tabulate(combined_df, headers=['Rank', 'Calories', ' Protein (g) ', 'Cholesterol (mg)', ' Sugars (g)', 'Vitamins'], tablefmt=table_format, showindex=False))
```
